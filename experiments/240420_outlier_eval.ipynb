```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_labels = pd.read_csv("../strength_covariance/data_ays/label_dict.csv")
label_dict = df_labels.to_dict(orient="records")[0]
```

```python
df = pd.read_csv("../data/models_w_props.csv",index_col = 0)
print(df.columns)
```

```python
df = df[['c44_fcc','c44_bcc','c44_sc',
         'c11_fcc','c11_bcc','c11_sc',
         'c12_fcc','c12_bcc','c12_sc',
         'bulk_modulus_fcc','bulk_modulus_bcc','bulk_modulus_sc',
         'vacancy_migration_energy_fcc','vacancy_migration_energy_bcc','vacancy_migration_energy_sc',
         'extr_stack_fault_energy_fcc','intr_stack_fault_energy_fcc',
         'cohesive_energy_fcc','cohesive_energy_bcc','cohesive_energy_diamond']
         ]
df.columns = [label_dict[i] for i in df.columns]
df.columns
```

```python
ax = df.boxplot(['eSFE FCC',
                 'iSFE FCC'],
           figsize = (3,3),
           rot = 90)
ax.set_ylabel(r'ev/A$^2$')
ax.set_yscale('log')
plt.savefig("../figures/si/bp_sfe.pdf",bbox_inches = 'tight')
plt.show()
```

```python
ax = df.boxplot(['Coh. Energy FCC',
                 'Coh. Energy BCC',
                 'Coh. Energy Diamond'],
           figsize = (3,3),
           rot = 90)
ax.set_ylabel(r'ev/A$^2$')
ax.set_yscale('log')
plt.savefig("../figures/si/bp_cohEn.pdf",bbox_inches = 'tight')
plt.show()
```

```python
ax = df.boxplot(['C44 FCC','C44 BCC','C44 SC',
                 'C11 FCC','C11 BCC','C11 SC',
                 'C12 FCC','C12 BCC','C12 SC',
                 'Bulk Modulus FCC', 'Bulk Modulus BCC', 'Bulk Modulus SC'],
                 rot=90,
                 figsize = (5,3))
ax.set_ylabel('GPa')
ax.set_yscale('log')
plt.savefig("../figures/si/bp_ec.pdf",bbox_inches = 'tight')
plt.show()
```
